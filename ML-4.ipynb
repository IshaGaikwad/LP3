```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(-10,10)
Y = (X+3)**2
print(X,Y)
```

```python
plt.plot(X,Y,color="black")
plt.xlabel = ("X")
plt.ylabel = ("F(X)")
plt.show()
```

```python
x = 2
lr = 0.1                   #learning rate
precision = 0.001          #accuracy
change_in_x = 1
itr = 0
min = 0

plt.plot(X,Y,color="black")

while precision < change_in_x and itr < 1000:
    grad = 2*(x+3)
    prev_x = x
    x = x - lr * grad
    itr += 1
    change_in_x = abs(x - prev_x)
    y = (x+3)**2
    plt.scatter(x,y)
    min = x
    print(itr, "\t", x)
```
